```python
import pandas as pd
```

```python
df= pd.read_csv('survey_results_public.csv')
```

```python
df.head()
```

```python
men_count=df["Gender"].value_counts()["Man"]
```

```python
no_men=df[df["Gender"]!="Man"]
no_men.head()
```

```python
df.loc[df["Age1stCode"] == "Younger than 5 years", "Age1stCode"] = "0 - 4 years"
```

```python
df.loc[df["Age1stCode"] == "Older than 64 years", "Age1stCode"] = "64 years +"
```

```python
df.loc[df["Age1stCode"] == "5 - 10 years", "Age1stCode"] = "05 - 10 years"
```

```python
df.dropna(subset=["Age1stCode", "SOPartFreq"])
```

```python
df.to_csv('allGenders.csv')
```

```python
cleaned= no_men[no_men["Gender"]!="Prefer not to say"]
cleaned.head(20)
```

```python

no_NaNs=cleaned.dropna(subset=["Gender"])
```

```python
def count_nb(df_col):
    count=0
    for item in df_col:
        if "Non-binary" in item:
            count+=1
    return count

```

```python
no_NaNs.to_csv('genderMinorities.csv')
```

```python
gender_col=list(new_df["Gender"])
```

```python
nb_count= count_nb(gender_col)
```

```python
women_count=new_df["Gender"].value_counts()["Woman"]
trans_count=new_df["Trans"].value_counts()["Yes"]
```

```python
gender_count_dict={"Gender":["Men","Women", "Trans", "Non-binary/Genderqueer"],
"Count":[men_count, women_count,trans_count, nb_count]}
```

```python
df_counts=pd.DataFrame.from_dict(gender_count_dict)
```

```python
df_counts.to_csv("allgendersCount.csv",  index=False)
```
